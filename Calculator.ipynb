```python
def average_score(english,maths,urdu):
  average_score= (english+maths+urdu)/3
  return(average_score)
english =int(input("what is your english score?"))
maths =int(input("what is your maths score?"))
urdu =int(input("what is your urdu score?"))
calculated_avg = average_score(english,maths,urdu)
if calculated_avg >=90:
   grade ="A+"
elif calculated_avg >=80:
  grade ="B+"
elif calculated_avg >=70:
  grade ="C+"
elif calculated_avg >= 60:
  grade ="D+"
else:
  grade = "F"
print(f" your average score is {calculated_avg:.2f} and your grade is {grade} ")
```
